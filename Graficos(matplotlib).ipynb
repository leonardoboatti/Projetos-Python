```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
dados = pd.read_csv('extra/dados/aluguel.csv', sep =';')
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor X Área')

g2.hist(dados.Valor)
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valor = grupo.mean().values
g4.bar(label, valor)
g4.set_title('Valor Médio por Tipo')

area
```

```python
area.savefig('grafico-correto.png', dpi = 300, bbox_inches = 'tight')
```

```python
dados_ex = pd.read_csv('dados/aluguel_amostra.csv', sep = ';')
```

```python
area_ex = plt.figure()
```

```python
g1_ex = area_ex.add_subplot(1, 2, 1)
g2_ex = area_ex.add_subplot(1, 2, 2)

grupo1 = dados_ex.groupby('Tipo Agregado')['Valor']
label_ex = grupo1.count().index
valores_ex = grupo1.count().values
g1_ex.pie(valores_ex, labels = label_ex, autopct='%1.1f%%')
g1_ex.set_title('Total de Imóveis por Tipo Agregado')

grupo2 = dados_ex.groupby('Tipo')['Valor']
label_ex2 = grupo2.count().index
valores_ex2 = grupo2.count().values
g2_ex.pie(valores_ex2, labels = label_ex2, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2_ex.set_title('Total de Imóveis por Tipo')

area_ex
```
